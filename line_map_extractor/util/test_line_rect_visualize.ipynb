```python
import numpy as np
import matplotlib.pyplot as plt
from check_line_rect import check_line_rect
```

```python
def draw_test_case(p1, p2, w, h, title, ax):
    """
    Draw a single test case showing the rectangle and line.
    
    Args:
        p1: First endpoint of the line
        p2: Second endpoint of the line
        w: Rectangle width
        h: Rectangle height
        title: Title for the subplot
        ax: Matplotlib axis to draw on
    """
    # Draw rectangle
    rect = plt.Rectangle((0, 0), w, h, fill=False, edgecolor='black', linewidth=2, label='Rectangle')
    ax.add_patch(rect)
    
    # Draw line
    ax.plot([p1[0], p2[0]], [p1[1], p2[1]], 'r-', linewidth=2, label='Line')
    
    # Mark endpoints
    ax.plot(p1[0], p1[1], 'bo', markersize=8, label='Endpoint 1')
    ax.plot(p2[0], p2[1], 'go', markersize=8, label='Endpoint 2')
    
    # Test intersection
    result = check_line_rect(p1, p2, w, h)
    
    # Set title with result
    status = "INTERSECTS" if result else "NO INTERSECTION"
    ax.set_title(f"{title}\n{status}", fontsize=12, fontweight='bold')
    
    # Set labels and grid
    ax.set_xlabel('X coordinate')
    ax.set_ylabel('Y coordinate')
    ax.grid(True, alpha=0.3)
    ax.legend()
    
    # Set equal aspect ratio and limits
    ax.set_aspect('equal')
    margin = 50
    ax.set_xlim(min(p1[0], p2[0], 0) - margin, max(p1[0], p2[0], w) + margin)
    ax.set_ylim(min(p1[1], p2[1], 0) - margin, max(p1[1], p2[1], h) + margin)

def create_visualization():
    """Create a comprehensive visualization of all test cases."""
    
    # Define rectangle dimensions
    w, h = 200, 200
    
    # Define test cases
    test_cases = [
        {
            'p1': np.array([50, 50]),
            'p2': np.array([100, 100]),
            'title': 'Test 1: Line Inside Rectangle'
        },
        {
            'p1': np.array([300, 300]),
            'p2': np.array([400, 400]),
            'title': 'Test 2: Line Outside Rectangle'
        },
        {
            'p1': np.array([-50, 100]),
            'p2': np.array([250, 100]),
            'title': 'Test 3: Line Crossing Rectangle'
        },
        {
            'p1': np.array([0, 50]),
            'p2': np.array([0, 150]),
            'title': 'Test 4: Line Touching Edge'
        },
        {
            'p1': np.array([-100, 50]),
            'p2': np.array([300, 50]),
            'title': 'Test 5: Horizontal Line Crossing'
        },
        {
            'p1': np.array([100, -100]),
            'p2': np.array([100, 300]),
            'title': 'Test 6: Vertical Line Crossing'
        },
        {
            'p1': np.array([-50, -50]),
            'p2': np.array([250, 250]),
            'title': 'Test 7: Diagonal Line Crossing'
        },
        {
            'p1': np.array([200, 100]),
            'p2': np.array([300, 100]),
            'title': 'Test 8: Line Starting at Edge'
        }
    ]
    
    # Create subplots
    fig, axes = plt.subplots(2, 4, figsize=(16, 10))
    fig.suptitle('Line-Rectangle Intersection Test Cases', fontsize=16, fontweight='bold')
    
    # Flatten axes for easier iteration
    axes = axes.flatten()
    
    # Draw each test case
    for i, test_case in enumerate(test_cases):
        draw_test_case(
            test_case['p1'], 
            test_case['p2'], 
            w, h, 
            test_case['title'], 
            axes[i]
        )
    
    plt.tight_layout()
    plt.show()
```

```python
def create_single_test_visualization():
    """Create a detailed visualization of a single test case."""
    
    w, h = 200, 200
    
    # Test case: Line crossing rectangle
    p1 = np.array([-50, 100])
    p2 = np.array([250, 100])
    
    fig, ax = plt.subplots(1, 1, figsize=(10, 8))
    
    # Draw rectangle
    rect = plt.Rectangle((0, 0), w, h, fill=False, edgecolor='black', linewidth=3, label='Rectangle')
    ax.add_patch(rect)
    
    # Draw line
    ax.plot([p1[0], p2[0]], [p1[1], p2[1]], 'r-', linewidth=3, label='Line Segment')
    
    # Mark endpoints
    ax.plot(p1[0], p1[1], 'bo', markersize=10, label=f'P1 ({p1[0]}, {p1[1]})')
    ax.plot(p2[0], p2[1], 'go', markersize=10, label=f'P2 ({p2[0]}, {p2[1]})')
    
    # Mark intersection points
    intersection_points = [(0, 100), (200, 100)]
    for i, point in enumerate(intersection_points):
        ax.plot(point[0], point[1], 'ro', markersize=8, label=f'Intersection {i+1}')
    
    # Test intersection
    result = check_line_rect(p1, p2, w, h)
    
    ax.set_title(f'Line-Rectangle Intersection Example\nResult: {"INTERSECTS" if result else "NO INTERSECTION"}', 
                 fontsize=14, fontweight='bold')
    ax.set_xlabel('X coordinate')
    ax.set_ylabel('Y coordinate')
    ax.grid(True, alpha=0.3)
    ax.legend()
    ax.set_aspect('equal')
    
    # Set limits to show the full context
    margin = 50
    ax.set_xlim(min(p1[0], p2[0], 0) - margin, max(p1[0], p2[0], w) + margin)
    ax.set_ylim(min(p1[1], p2[1], 0) - margin, max(p1[1], p2[1], h) + margin)
    
    plt.tight_layout()
    plt.show()
```

```python
print("Creating visualizations for line-rectangle intersection test cases...")

# Create comprehensive visualization
print("1. Creating comprehensive test case visualization...")
create_visualization()

# Create detailed single test case
print("2. Creating detailed single test case visualization...")
create_single_test_visualization()

print("Visualizations complete!") 
```
